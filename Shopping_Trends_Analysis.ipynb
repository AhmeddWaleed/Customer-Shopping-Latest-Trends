### Analyzing and Visualizing Data

```python
df['Customer ID'].nunique()
```

```python
# Total Spent by Customers

df.groupby("Customer ID")["Purchase Amount (USD)"].sum().sort_values(ascending=False)
```

```python
# Gender Distribution

Gender_Distribution = df["Gender"].value_counts()
Gender_Distribution
```

```python
plt.pie(Gender_Distribution, labels=Gender_Distribution.index , autopct='%.1f%%', startangle=90 , colors=['#727bf7' , '#da87ed']  )
plt.title("Gender Distribution")
plt.legend(labels=Gender_Distribution.index ,loc='upper right' )
plt.show()
```

```python
# Subscription Distribution 

Subscription_Distribution = df["Subscription Status"].value_counts()
Subscription_Distribution
```

```python
plt.pie(Subscription_Distribution, labels=Subscription_Distribution.index, startangle=90 , colors=['#faae57' , '#9ef763'], autopct='%.1f%%')
plt.title('Subscription Distribution')
plt.legend(labels=Subscription_Distribution.index , loc='upper right')
plt.show()
```

```python
# Customer Count by Gender and Subscription Status

df.pivot_table(index="Gender",columns="Subscription Status",values="Customer ID",aggfunc="count",fill_value=0)
```

```python
# Total Sales by Category

Total_Sales_by_Category = df.groupby("Category")["Purchase Amount (USD)"].sum().sort_values( ascending=False)
Total_Sales_by_Category
```

```python
plt.pie(Total_Sales_by_Category , labels=Total_Sales_by_Category.index , startangle=90 , autopct='%.1f%%' 
        , colors = [ '#3593f2' , '#57a8fa'  , '#3cc9f0' , '#84edf5'])
plt.legend( labels=Total_Sales_by_Category.index , loc='upper right', bbox_to_anchor=(1.5, 1))
plt.title('Total Sales by Category')
plt.show()
```

```python
# Quantity by Category

Quantity_by_Category = df["Category"].value_counts().sort_values(ascending=False)
Quantity_by_Category
```

```python
fig, ax = plt.subplots(figsize=(5,5))
bars =plt.bar(Quantity_by_Category.index , Quantity_by_Category.values , color= '#95f542' , width=0.4)
plt.title('Quantity by Category')
plt.xlabel('Category')
plt.ylabel('Quantity')
ax.spines[['right' , 'top']].set_visible(False)
plt.bar_label(bars, fmt='%.0f', padding=3)
plt.show()
```

```python
# Total Sales by Items

Total_Sales_by_Items = df.groupby("Item Purchased")["Purchase Amount (USD)"].sum().sort_values(ascending=False)
Total_Sales_by_Items
```

```python
fig, ax = plt.subplots(figsize=(8,8))
bars = plt.barh(Total_Sales_by_Items.index, Total_Sales_by_Items.values, color='#5da9f5')
plt.ylabel("Items")
plt.xlabel('Total Sales')
plt.title("Total Sales by Item")
ax.spines[['right' , 'top']].set_visible(False)
plt.bar_label(bars, fmt='%.0f', padding=3)
plt.show()
```

```python
# Quantity by Items

Quantity_by_Items = df["Item Purchased"].value_counts().sort_values(ascending=False)
Quantity_by_Items
```

```python
fig, ax = plt.subplots(figsize=(8,8))
bars = plt.barh(Quantity_by_Items.index, Quantity_by_Items.values, color='orange')
plt.ylabel("Items")
plt.xlabel("Quantity")
plt.title("Quantity by Items")
ax.spines[['right' , 'top']].set_visible(False)
plt.bar_label(bars, fmt='%.0f', padding=3)
plt.show()
```

```python
# Total Sales by Category and Items

df.groupby(["Category" , "Item Purchased"])["Purchase Amount (USD)"].sum().sort_values(ascending=False)
```

```python
# Total Sales by Location

Total_Sales_by_Location = df.groupby("Location")["Purchase Amount (USD)"].sum().sort_values(ascending=False)
Total_Sales_by_Location
```

```python
fig, ax = plt.subplots(figsize=(8,10))
bars = plt.barh(Total_Sales_by_Location.index, Total_Sales_by_Location.values, color='#5ad5fa')
plt.ylabel("Location")
plt.xlabel("Total Sales")
plt.title("Total Sales by Location")
ax.spines[['right' , 'top']].set_visible(False)
plt.bar_label(bars, fmt='%.0f', padding=3)
plt.show()
```

```python
# Total Sales by Location and Category

df.pivot_table(index="Location", columns="Category", values="Purchase Amount (USD)", aggfunc="sum", fill_value=0 )
```

```python
# Customer Count by Location

Customer_Count_by_Location = df.groupby("Location")["Customer ID"].count().sort_values(ascending=False)
Customer_Count_by_Location
```

```python
fig , ax = plt.subplots(figsize=(8,10)) 
bars = plt.barh(Customer_Count_by_Location.index, Customer_Count_by_Location.values , color='#ba3bf5')
plt.xlabel("Location")
plt.ylabel("Customer Count")
plt.title("Customer Count by Location")
ax.spines[['right', 'top']].set_visible(False)
plt.bar_label(bars, fmt='%.0f', padding=3)
plt.show()
```

```python
# Total Sales by Season

Total_Sales_by_Season = df.groupby("Season")["Purchase Amount (USD)"].sum().sort_values(ascending=False)
Total_Sales_by_Season
```

```python
plt.pie(Total_Sales_by_Season, labels=Total_Sales_by_Season.index, startangle=90 ,colors=['#a8f57f' , '#f7b140', '#6cc7f5', '#c375f0'], autopct='%.1f%%')
plt.title("Total Sales by Season")
plt.legend(labels=Total_Sales_by_Season.index , loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
# Customer Count by Season

Customer_Count_by_Season = df.groupby("Season")["Customer ID"].count().sort_values(ascending=False)
Customer_Count_by_Season
```

```python
plt.pie(Customer_Count_by_Season, labels=Customer_Count_by_Season.index, startangle=90 , autopct='%.1f%%' 
        , colors = [ '#3593f2' , '#57a8fa'  , '#3cc9f0' , '#84edf5'])
plt.legend(labels=Customer_Count_by_Season.index , loc='upper right', bbox_to_anchor=(1.3,1))
plt.title("Customer Count by Season")
plt.show()
```

```python
# Average Rating by Items

df.groupby("Item Purchased")["Review Rating"].mean().sort_values(ascending=False)
```

```python
# Creating new column containing rating classification

df.loc[df["Review Rating"] <= 2.5, 'Level'] = 'Low'
df.loc[df["Review Rating"] > 2.5 , 'Level'] = 'Medium'
df.loc[df["Review Rating"] > 3.7 , 'Level'] = 'High'
```
